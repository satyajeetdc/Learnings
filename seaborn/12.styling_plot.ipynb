```python
import seaborn as sns
from matplotlib import pyplot as plt
import pandas as pd
```

```python
var = pd.read_csv("tips.csv")
var.head()
```

```python
sns.boxplot(y = "total_bill", data = var)
plt.show()
```

```python
sns.barplot(x = "day", y = "total_bill", data = var)
plt.grid()
plt.show()
```

```python
sns.set_style("dark")
sns.barplot(x = "day", y = "total_bill", data = var)
plt.show()
```

```python
sns.set_style("whitegrid")
sns.barplot(x = "day", y = "total_bill", data = var)
plt.show()
```

```python
sns.set_style("darkgrid")
sns.barplot(x = "day", y = "total_bill", data = var)
plt.show()
```

```python
sns.set_style("ticks")
sns.barplot(x = "day", y = "total_bill", data = var)
plt.show()
```

```python
sns.set_style("whitegrid")
sns.barplot(x = "day", y = "total_bill", data = var)
sns.despine()
plt.show()
```

```python
sns.set_style("whitegrid")
plt.figure(figsize =(2,2))
sns.barplot(x = "day", y = "total_bill", data = var)
plt.show()
```

```python
sns.set_style("whitegrid")
plt.figure(figsize =(20,5))
sns.barplot(x = "day", y = "total_bill", data = var)
plt.show()
```

```python
sns.set_style("white")
sns.set_context("poster",font_scale = 0.8)
sns.barplot(x = "day", y = "total_bill", data = var)
plt.show()
```

```python
sns.set_style("white")
sns.set_context("paper",font_scale = 1)
sns.barplot(x = "day", y = "total_bill", data = var)
plt.show()
```
