```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
var = sns.load_dataset("penguins")
var.head()
```

```python
sns.displot(var["bill_length_mm"])
plt.show()
```

```python
sns.displot(var["flipper_length_mm"])
plt.show()
```

```python
sns.displot(var["flipper_length_mm"], bins = [170,180,190,200,210,220,230,240])
plt.show()
```

```python
sns.displot(var["flipper_length_mm"], bins = [170,180,190,200,210,220,230,240], kde = True)
plt.show()
```

```python
sns.displot(var["flipper_length_mm"], bins = [170,180,190,200,210,220,230,240], kde = True, rug = True)
plt.show()
```

```python
sns.displot(var["flipper_length_mm"], bins = [170,180,190,200,210,220,230,240], kde = True, rug = True, color = "y")
plt.show()
```

```python
sns.displot(var["flipper_length_mm"], rug = True, log_scale = True)
plt.show()
```
