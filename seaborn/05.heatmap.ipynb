```python
var1 = np.linspace(1,10,10).reshape(2,5)
var1
```

```python
sns.heatmap(var1,cmap = "PuOr", annot = True)
plt.show()
```

```python
ar =np.array([["a0","a1","a2","a3","a4"],["b0","b1","b2","b3","b4"]])

sns.heatmap(var1,cmap = "PuOr", annot = ar, fmt = "s")
plt.show()
```

```python
y = {"fontsize" : 15, "color" : "r"}
sns.heatmap(x, cmap = "PuOr", annot = True, annot_kws = y)
plt.show()
```

```python
y = {"fontsize" : 12, "color" : "green"}
sns.heatmap(x, cmap = "gist_heat", annot = True, annot_kws = y, linewidths = 10)
plt.show()
```

```python
y = {"fontsize" : 12, "color" : "green"}
sns.heatmap(x, cmap = "gist_heat", annot = True, annot_kws = y, linewidths = 10, linecolor = "b")
plt.show()
```

```python
y = {"fontsize" : 12, "color" : "green"}
sns.heatmap(x, cmap = "gist_heat", annot = True, annot_kws = y, linewidths = 10, linecolor="b", cbar = False)
plt.show()
```

```python
y = {"fontsize" : 12, "color" : "green"}
sns.heatmap(x, cmap = "gist_heat", annot = True, annot_kws = y, linewidths = 10, xticklabels = False, yticklabels = False)
plt.show()
```

```python
y = {"fontsize" : 12, "color" : "green"}
v = sns.heatmap(x, cmap = "gist_heat", annot = True, annot_kws = y, linewidths = 10)
v.set(xlabel = "Label1", ylabel = "Label2")
plt.show()
```

```python
y = {"fontsize" : 12, "color" : "green"}
v = sns.heatmap(x, cmap = "gist_heat", annot = True, annot_kws = y, linewidths = 10)
v.set(xlabel = "Label1", ylabel = "Label2")
sns.set(font_scale = 1.3)
plt.show()
```
