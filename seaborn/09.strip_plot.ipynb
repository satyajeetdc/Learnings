```python
import seaborn as sns
from matplotlib import pyplot as plt
import pandas as pd
```

```python
var = pd.read_csv("tips.csv")
var.head()
```

```python
sns.stripplot(x = "day", y = "total_bill", data = var)
plt.show()
```

```python
sns.stripplot(x = "day", y = "total_bill", data = var, hue = "sex")
plt.show()
```

```python
sns.stripplot(x = "day", y = "total_bill", data = var, hue = "sex", palette = "rocket")
plt.show()
```

```python
sns.stripplot(x = "total_bill", y = "day", data = var, hue = "sex", palette = "rocket")
plt.show()
```

```python
sns.stripplot(x = "day", y = "total_bill", data = var, hue = "sex", linewidth = 1.2, edgecolor = "black")
plt.show()
```

```python
sns.stripplot(x = "day", y = "total_bill", data = var, hue = "sex", linewidth = 1.2, edgecolor = "black", jitter = 0.3)
plt.show()
```

```python
sns.stripplot(x = "day", y = "total_bill", data = var, hue = "sex", size = 7)
plt.show()
```

```python
sns.stripplot(x = "day", y = "total_bill", data = var, hue = "sex", marker = "*", size = 10)
plt.show()
```

```python
sns.stripplot(x = "day", y = "total_bill", data = var, hue = "sex", marker = "*", size = 10, alpha = 0.6)
plt.show()
```

```python
sns.stripplot(y = var["total_bill"], data = var, hue = "sex")
plt.show()
```
