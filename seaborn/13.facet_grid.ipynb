```python
import seaborn as sns
from matplotlib import pyplot as plt
import pandas as pd
```

```python
var = pd.read_csv("tips.csv")
var.head()
```

```python
sns.boxplot(x = "tip", y = "day", data = var, hue = "sex")
plt.show()
```

```python
fg = sns.FacetGrid(var, col = "sex", hue = "day")
fg.map(plt.scatter, "total_bill", "tip").add_legend()
plt.show()
```

```python
fg = sns.FacetGrid(var, col = "day", hue = "sex")
fg.map(plt.scatter, "total_bill", "tip").add_legend()
plt.show()
```

```python
fg = sns.FacetGrid(var, col = "sex", hue = "day")
fg.map(plt.bar, "total_bill", "tip").add_legend()
plt.show()
```

```python
fg = sns.FacetGrid(var, col = "day", hue = "sex", palette = "turbo")
fg.map(plt.scatter, "total_bill", "tip").add_legend()
plt.show()
```

```python
fg = sns.FacetGrid(var, col = "day", hue = "sex", palette = "turbo")
fg.map(plt.scatter, "total_bill", "tip", edgecolor = "r").add_legend()
plt.show()
```
