```python
import seaborn as sns
from matplotlib import pyplot as plt
import pandas as pd
```

```python
var = pd.read_csv("tips.csv")
var.head()
```

```python
sns.catplot(x = "size", y = "tip", data = var)
plt.show()
```

```python
sns.catplot(x = "size", y = "tip", data = var, kind = "point")
plt.show()
```

```python
sns.catplot(x = "size", y = "tip", data = var, kind = "point", hue = "sex")
plt.show()
```

```python
sns.catplot(x = "size", y = "tip", data = var, kind = "box", hue = "sex")
plt.show()
```

```python
sns.catplot(x = "size", y = "tip", data = var, kind = "swarm", hue = "sex")
plt.show()
```

```python
sns.catplot(x = "size", y = "tip", data = var, kind = "boxen", hue = "sex")
plt.show()
```

```python
sns.catplot(x = "size", y = "tip", data = var, kind = "violin", hue = "sex")
plt.show()
```

```python
sns.catplot(x = "size", y = "tip", data = var, kind = "bar")
plt.show()
```
