```python
import seaborn as sns
from matplotlib import pyplot as plt
```

```python
var = [1,2,3,4,5,6,7]
var1= [2,3,4,1,5,6,7]
```

```python
plt.plot(var, var1)
plt.show()
```

```python
import pandas as pd
```

```python
x1 = pd.DataFrame({"var" : var, "var1" : var1} )

sns.lineplot(x = "var", y = "var1", data = x1)
plt.show()
```

```python
y1 = sns.load_dataset("penguins")
```

```python
y1.head()
```

```python
sns.lineplot(x = "bill_length_mm", y = "bill_depth_mm", data = y1)
plt.show()
```

```python
sns.lineplot(x = "bill_length_mm", y = "bill_depth_mm", data = y1, hue = "sex", style = "sex")
plt.show()
```

```python
sns.lineplot(x = "bill_length_mm", y = "bill_depth_mm", data = y1, hue = "sex", size = 50)
plt.show()
```

```python
sns.lineplot(x = "bill_length_mm", y = "bill_depth_mm", data = y1, hue = "sex", style = "sex", palette = "turbo")
plt.show()
```

```python
y2 = sns.load_dataset("penguins")
```

```python
y3 = y2.head(20)
```

```python
sns.lineplot(x = "bill_length_mm", y = "bill_depth_mm", data = y3, hue ="sex", style = "sex" , markers = ["o","<"])
plt.show()
```

```python
sns.lineplot(x = "bill_length_mm", y = "bill_depth_mm", data = y3, hue ="sex", style = "sex" , 
             markers = ["o","<"], dashes = False)
plt.show()
```

```python
sns.lineplot(x = "bill_length_mm", y = "bill_depth_mm", data = y3, hue ="sex", style = "sex" , 
             markers = ["o","<"], dashes = False, legend = False)
plt.show()
```

```python
sns.lineplot(x = "bill_length_mm", y = "bill_depth_mm", data = y3, hue ="sex", style = "sex" , 
             markers = ["o","<"])
plt.grid()
plt.title("penguins")
plt.show()
```
