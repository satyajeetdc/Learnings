```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
var = sns.load_dataset("penguins")
```

```python
var.head()
```

```python
sns.barplot(x = var.island, y = var.bill_length_mm)
plt.show()
```

```python
var["island"].value_counts()
```

```python
sns.barplot(x = "island", y = "bill_length_mm", data = var)
plt.show()
```

```python
sns.barplot(x = "island", y = "bill_length_mm", data = var, hue = "sex")
plt.show()
```

```python
order_1 = ["Dream", "Torgersen", "Biscoe"]
sns.barplot(x = "island", y = "bill_length_mm", data = var, hue = "sex", order = order_1)
plt.show()
```

```python
order_1 = ["Dream", "Torgersen", "Biscoe"]
sns.barplot(x = "island", y = "bill_length_mm", data = var, hue = "sex", order = order_1, hue_order = ["Female","Male"])
plt.show()
```

```python
sns.barplot(x = "island", y = "bill_length_mm", data = var, hue = "sex", ci = 0.95)
plt.show()
```

```python
sns.barplot(x = "island", y = "bill_length_mm", data = var, hue = "sex", n_boot = 3)
plt.show()
```

```python
sns.barplot(x = "bill_length_mm", y = "bill_depth_mm", data = var, hue = "sex")
plt.show()
```

```python
sns.barplot(x = "bill_length_mm", y = "bill_depth_mm", data = var, hue = "sex", orient = "h")
plt.show()
```

```python
sns.barplot(x = "island", y = "bill_length_mm", data = var, hue = "sex", hue_order = ["Female","Male"], palette = "icefire")
plt.show()
```

```python
sns.barplot(x = "island", y = "bill_length_mm", data = var, hue = "sex", hue_order = ["Female","Male"], 
            palette = "icefire", saturation = 0.1)
plt.show()
```

```python
sns.barplot(x = "island", y = "bill_length_mm", data = var, hue = "sex", hue_order = ["Female","Male"], 
            palette = "icefire", errcolor = "r", errwidth = 12)
plt.show()
```

```python
sns.barplot(x = "island", y = "bill_length_mm", data = var, hue = "sex", hue_order = ["Female","Male"], 
            capsize = 0.1)
plt.show()
```

```python
sns.barplot(x = "island", y = "bill_length_mm", data = var, hue = "sex", hue_order = ["Female","Male"], 
            dodge = False)
plt.show()
```

```python
sns.barplot(x = "island", y = "bill_length_mm", data = var, hue = "sex", hue_order = ["Female","Male"], 
            alpha = 0.5)
plt.show()
```

```python
sns.set(style = "dark")
sns.barplot(x = "island", y = "bill_length_mm", data = var, hue = "sex", hue_order = ["Female","Male"])
plt.show()
```
