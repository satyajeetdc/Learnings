```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
var = sns.load_dataset("penguins")
var.head()
```

```python
sns.scatterplot(x = "bill_length_mm", y = "bill_depth_mm", data = var)
plt.show()
```

```python
sns.scatterplot(x = "bill_length_mm", y = "flipper_length_mm", data = var)
plt.show()
```

```python
sns.scatterplot(x = "bill_length_mm", y = "flipper_length_mm", data = var, hue = "sex", style = "sex")
plt.show()
```

```python
sns.scatterplot(x = "bill_length_mm", y = "flipper_length_mm", data = var, hue = "sex", style = "sex", palette = "rocket")
plt.show()
```

```python
sns.scatterplot(x = "bill_length_mm", y = "flipper_length_mm", data = var, hue = "sex", style = "sex", 
                size = "sex", sizes = (100,100), palette = "rocket")
plt.show()
```

```python
sns.scatterplot(x = "bill_length_mm", y = "flipper_length_mm", data = var, hue = "sex", style = "sex", 
                size = "sex", sizes = (100,100), palette = "rocket", alpha = 0.6)
plt.show()
```

```python
sns.scatterplot(x = "bill_length_mm", y = "flipper_length_mm", data = var, hue = "sex", style = "sex", 
                size = "sex", sizes = (100,100), palette = "rocket", markers = [">","*"])
plt.show()
```

```python
m = {"Male" : ">", "Female" : "*"}
sns.scatterplot(x = "bill_length_mm", y = "flipper_length_mm", data = var, hue = "sex", style = "sex", 
                size = "sex", sizes = (100,100), palette = "rocket", markers = m)
plt.show()
```
