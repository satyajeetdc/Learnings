```python
import seaborn as sns
from matplotlib import pyplot as plt
import pandas as pd
```

```python
var = pd.read_csv("tips.csv")
var.head()
```

```python
sns.set(style = "whitegrid")
sns.boxplot(x = "day", y = "total_bill", data = var)
plt.show()
```

```python
sns.boxplot(x = "day", y = "total_bill", data = var)
plt.show()
```

```python
sns.boxplot(y = "day", x = "total_bill", data = var, hue = "time")
plt.show()
```

```python
sns.boxplot(x = "day", y = "total_bill", data = var, palette = "rocket")
plt.show()
```

```python
sns.boxplot(x = "total_bill", y = "day", data = var, palette = "rocket", order = ["Thur","Fri","Sat","Sun"])
plt.show()
```

```python
sns.boxplot(x = "day", y = "total_bill", data = var, palette = "rocket", showmeans = True, 
            meanprops = {"marker" : "o", "markeredgecolor" : "r", "markerfacecolor" : "white"})
plt.show()
```

```python
sns.boxplot(x = "day", y = "total_bill", data = var, palette = "plasma", linewidth = 3)
plt.show()
```

```python
sns.boxplot(x = "tip", y = "total_bill", data = var, palette = "brg", orient = "h")
plt.show()

# "h" only works when both x and y axis are numeric data
```

```python
sns.boxplot(y = "total_bill", data = var)
plt.show()
```
