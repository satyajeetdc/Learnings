```python
import seaborn as sns
from matplotlib import pyplot as plt
import pandas as pd
```

```python
var = pd.read_csv("tips.csv")
var.head()
```

```python
sns.pairplot(var)
plt.show()
```

```python
sns.pairplot(var, hue ="sex")
plt.show()
```

```python
sns.pairplot(var, vars = ["total_bill", "tip"], hue ="sex")
plt.show()
```

```python
sns.pairplot(var, vars = ["total_bill", "tip"], hue ="sex", hue_order = ["Male","Female"], palette = "rocket")
plt.show()
```

```python
sns.pairplot(var, hue ="sex", hue_order = ["Male","Female"], x_vars = ["total_bill","tip"], palette = "rocket")
plt.show()
```

```python
sns.pairplot(var, hue ="sex", hue_order = ["Male","Female"], y_vars = ["total_bill","tip"], palette = "rocket")
plt.show()
```

```python
sns.pairplot(var, hue ="sex", hue_order = ["Male","Female"], kind = "reg", palette = "rocket")
plt.show()
```

```python
sns.pairplot(var, hue ="sex", hue_order = ["Male","Female"], kind = "scatter", palette = "rocket")
plt.show()
```

```python
sns.pairplot(var, hue ="sex", hue_order = ["Male","Female"], kind = "kde")
plt.show()
```

```python
sns.pairplot(var, hue ="sex", hue_order = ["Male","Female"], kind = "hist")
plt.show()
```

```python
sns.pairplot(var, hue ="sex", hue_order = ["Male","Female"], kind = "scatter", diag_kind = "kde", palette = "rocket")
plt.show()
```

```python
sns.pairplot(var, hue ="sex", hue_order = ["Male","Female"], diag_kind = "kde", markers = ["*",">"])
plt.show()
```
