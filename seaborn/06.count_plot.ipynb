```python
import seaborn as sns
from matplotlib import pyplot as plt
import pandas as pd
```

```python
var = pd.read_csv("tips.csv")
var.head()
```

```python
sns.countplot(x = "sex", data = var)
plt.show()
```

```python
sns.barplot(x = "sex", y = "size", data = var)
plt.show()
```

```python
sns.countplot(x = "sex", data = var, hue = "smoker")
plt.show()
```

```python
sns.countplot(y = "sex", data = var, hue = "smoker")
plt.show()
```

```python
sns.countplot(x = "sex", data = var, hue = "smoker", palette = "rainbow")
plt.show()
```

```python
sns.countplot(x = "sex", data = var, hue = "smoker", color = "m")
plt.show()
```
