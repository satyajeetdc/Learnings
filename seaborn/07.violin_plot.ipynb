```python
import seaborn as sns
from matplotlib import pyplot as plt
import pandas as pd
```

```python
var = pd.read_csv("tips.csv")
var.head()
```

```python
sns.violinplot(x = "day", y = "total_bill", data = var)
plt.show()
```

```python
sns.violinplot(x = "day", y = "total_bill", data = var, hue = "sex")
plt.show()
```

```python
sns.violinplot(x = "day", y = "total_bill", data = var, hue = "time")
plt.show()
```

```python
sns.violinplot(x = "day", y = "total_bill", data = var, linewidth = 4)
plt.show()
```

```python
sns.violinplot(x = "day", y = "total_bill", data = var, linewidth = 2, palette = "terrain_r")
plt.show()
```

```python
sns.violinplot(x = "time", y = "total_bill", data = var, linewidth = 2, palette = "rocket", order = ["Dinner","Lunch"])
plt.show()
```

```python
sns.violinplot(x = "total_bill", y = "time", data = var, linewidth = 2, palette = "rocket", 
               order = ["Dinner","Lunch"], orient = "h")
plt.show()
```

```python
sns.violinplot(x = "day", y = "total_bill", data = var, linewidth = 2, palette = "terrain_r", saturation = 0.5)
plt.show()
```

```python
sns.violinplot(x = "day", y = "total_bill", data = var, linewidth = 2, color = "m")
plt.show()
```

```python
sns.violinplot(x = "day", y = "total_bill", data = var, hue = "sex", split = True)
plt.show()
```

```python
sns.violinplot(x = "day", y = "total_bill", data = var, hue = "sex", split = True, scale = "count")
plt.show()
```
