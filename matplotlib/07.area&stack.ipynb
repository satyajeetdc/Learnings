```python
from matplotlib import pyplot as plt
```

```python
x = [1,2,3,5,8]
area = [7,2,3,6,8]
```

```python
plt.stackplot(x, area)
plt.show()
```

```python
x = [1,2,3,5,8]
area1 = [1,2,3,6,8]
area2 = [1,2,3,6,8]
area3 = [3,2,3,6,10]
plt.stackplot(x, area1, area2, area3)
plt.show()
```

```python
x = [1,2,3,5,8]
area1 = [1,2,3,6,8]
area2 = [1,2,3,6,8]
area3 = [3,2,3,6,10]
l = ["Area1", "Area2", "Area3"]
plt.stackplot(x, area1, area2, area3, labels = l)
plt.legend(loc = 2)
plt.show()
```

```python
x = [1,2,3,5,8]
area1 = [1,2,3,6,8]
area2 = [1,2,3,6,8]
area3 = [3,2,3,6,10]
l = ["Area1", "Area2", "Area3"]
c = ["y","g","m"]
plt.stackplot(x, area1, area2, area3, labels = l, colors = c)
plt.legend(loc = 2)
plt.show()
```

```python
x = [1,2,3,5,8]
area1 = [1,2,3,6,8]
area2 = [1,2,3,6,8]
area3 = [3,2,3,6,10]
l = ["Area1", "Area2", "Area3"]
c = ["y","g","m"]
plt.stackplot(x, area1, area2, area3, labels = l, colors = c, baseline = "zero")
plt.legend(loc = 2)
plt.show()
```

```python
x = [1,2,3,5,8]
area1 = [1,2,3,6,8]
area2 = [1,2,3,6,8]
area3 = [3,2,3,6,10]
l = ["Area1", "Area2", "Area3"]
c = ["y","g","m"]
plt.stackplot(x, area1, area2, area3, labels = l, colors = c, baseline = "sym")
plt.legend(loc = 2)
plt.show()
```

```python
x = [1,2,3,5,8]
area1 = [1,2,3,6,8]
area2 = [1,2,3,6,8]
area3 = [3,2,3,6,10]
l = ["Area1", "Area2", "Area3"]
c = ["y","g","m"]
plt.stackplot(x, area1, area2, area3, labels = l, colors = c, baseline = "wiggle")
plt.legend(loc = 2)
plt.show()
```

```python
x = [1,2,3,5,8]
area1 = [1,2,3,6,8]
area2 = [1,2,3,6,8]
area3 = [3,2,3,6,10]
l = ["Area1", "Area2", "Area3"]
c = ["y","g","m"]
plt.stackplot(x, area1, area2, area3, labels = l, colors = c)
plt.title("Sample")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
#plt.grid()
plt.legend(loc = 2)
plt.show()
```
