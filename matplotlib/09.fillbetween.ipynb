```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.array([1,2,3,4,5])
y = np.array([1,2,3,4,5])
```

```python
plt.plot(x, y, color = "r")

plt.fill_between(x, y)

plt.title("sample")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.show()
```

```python
plt.plot(x, y, color = "r")

plt.fill_between(x = [2,4], y1 = 2, y2 = 4)

plt.title("sample")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.show()
```

```python
plt.plot(x, y, color = "r")

plt.fill_between(x, y, color = "m", where = (x>=2) & (x<=4), alpha = 0.6)

plt.title("sample")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.show()
```
