```python
from matplotlib import pyplot as plt
```

```python
x = [10,20,20,30,40,60,100]
```

```python
plt.boxplot(x)
plt.show()
```

```python
plt.boxplot(x, vert = False)
plt.show()
```

```python
plt.boxplot(x, notch = True)
plt.show()
```

```python
plt.boxplot(x, widths = 0.25, vert = False)
plt.show()
```

```python
plt.boxplot(x, labels = ["sample"], patch_artist = True)
plt.show()
```

```python
plt.boxplot(x, labels = ["sample"], showmeans = True)
plt.show()
```

```python
plt.boxplot(x, labels = ["sample"], whis = 2)
plt.show()
```

```python
plt.boxplot(x, labels = ["sample"], sym = "r>")
plt.show()
```

```python
plt.boxplot(x, labels = ["sample"], showmeans = True, boxprops = {"color" : "red"}, sym = "r>",
            capprops = {"color" : "red"}, whiskerprops = {"color" : "g"}, flierprops = dict(markeredgecolor = "blue"))
plt.show()
```

```python
x = [20,50,100,180,250]
x1 = [20,30,40,100,200,150]
y = [x,x1]

plt.boxplot(y, labels=["Sample1","Sample2"],
            showmeans = True, boxprops = {"color" : "red"}, capprops = {"color" : "red"}, whiskerprops = {"color" : "g"})
plt.show()
```
