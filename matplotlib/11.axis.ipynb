```python
from matplotlib import pyplot as plt
```

```python
x = [1,2,3,4,5,6]
y = [10,8,2,6,8,12]
```

```python
plt.plot(x, y)
plt.show()
```

```python
plt.plot(x, y)

plt.xticks(x, labels = ["a","b","c","d","e","f"])
plt.yticks(y)

plt.show()
```

```python
plt.plot(x, y)

plt.xticks(x, labels = ["a","b","c","d","e","f"])
plt.yticks(y)
plt.xlim(0,10)
plt.ylim(0,20)

plt.show()
```

```python
plt.plot(x, y)

plt.axis([0,10,0,20])

plt.show()
```
