```python
from matplotlib import pyplot as plt
import numpy as np
import random
```

```python
x = np.random.randint(10,60,(50))
x
```

```python
nos = [33, 54, 55, 56, 54, 27, 34, 57, 50, 19, 31, 53, 50, 55, 34, 50, 50,
       55, 29, 16, 17, 27, 24, 49, 33, 36, 28, 39, 50, 47, 30, 14, 37, 30,
       14, 58, 30, 21, 23, 47, 31, 50, 57, 49, 52, 33, 38, 11, 10, 19]
```

```python
plt.hist(nos,color = "darkblue", edgecolor = "black", linewidth = 3)
plt.title("Histogram")
plt.xlabel("Python")
plt.ylabel("Range")
plt.show()
```

```python
l = [10,20,30,40,50,60]
plt.hist(nos,color = "violet", bins = l, edgecolor = "black", linewidth = 3)
plt.title("Histogram")
plt.xlabel("Python")
plt.ylabel("Range")
plt.show()
```

```python
l = [10,20,30,40,50,60]
plt.hist(nos,"auto", (0,100),edgecolor = "black")
plt.title("Histogram")
plt.xlabel("Python")
plt.ylabel("Range")
plt.show()
```

```python
plt.hist(nos,color = "darkblue", edgecolor = "black", linewidth = 3, cumulative = -1)
plt.title("Histogram")
plt.xlabel("Python")
plt.ylabel("Range")
plt.show()
```

```python
plt.hist(nos,color = "darkblue", edgecolor = "black", linewidth = 3, bottom = 20)
plt.title("Histogram")
plt.xlabel("Python")
plt.ylabel("Range")
plt.show()
```

```python
plt.hist(nos,color = "darkblue", edgecolor = "black", linewidth = 3, align = "mid")
plt.title("Histogram")
plt.xlabel("Python")
plt.ylabel("Range")
plt.show()
```

```python
plt.hist(nos,color = "darkblue", edgecolor = "red", linewidth = 3, histtype = "step")
plt.title("Histogram")
plt.xlabel("Python")
plt.ylabel("Range")
plt.show()
```

```python
plt.hist(nos,color = "darkblue", edgecolor = "black", linewidth = 3, histtype = "barstacked")
plt.title("Histogram")
plt.xlabel("Python")
plt.ylabel("Range")
plt.show()
```

```python
plt.hist(nos,color = "darkblue", edgecolor = "red", linewidth = 3, log = True)
plt.title("Histogram")
plt.xlabel("Python")
plt.ylabel("Range")
plt.show()
```

```python
plt.hist(nos,color = "darkblue", edgecolor = "red", linewidth = 3, label = "Python")
plt.title("Histogram")
plt.xlabel("Python")
plt.ylabel("Range")
plt.legend()
plt.show()
```

```python
plt.hist(nos,color = "darkblue", edgecolor = "red", linewidth = 3, label = "Python")
plt.title("Histogram")
plt.xlabel("Python")
plt.ylabel("Range")

plt.axvline(35, color = "yellow", linewidth = 3, label = "line")
plt.grid()

plt.legend()
plt.show()
```
