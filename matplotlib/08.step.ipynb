```python
from matplotlib import pyplot as plt
```

```python
import numpy as np
```

```python
x = np.array([1, 3, 4, 5, 7, 9])
y = np.array([1, 9, 16, 25, 49, 30])
```

```python
plt.step(x, y, marker = "o")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.show()
```

```python
plt.step(x, y, marker = "o", ms= 20, mfc = "g")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.show()
```

```python
plt.step(x, y, marker = "o", label = "Sample")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.legend(loc = 2)
plt.show()
```
