```python
from matplotlib import pyplot as plt
```

```python
day = [1,2,3,4,5,6,7]
no = [2,3,1,4,5,3,6]
c = ["yellow","red","blue","green","orange","brown","black"]
sizes = [200,300,450,350,600,400,300]
plt.scatter(day,no, color = c,s = sizes)
plt.title("Scatter Plot", fontsize = 12, font = "Times New Roman")
plt.xlabel("Day")
plt.ylabel("Number")
plt.show()
```

```python
day = [1,2,3,4,5,6,7]
no = [2,3,1,4,5,3,6]
c = ["yellow","red","blue","green","orange","brown","black"]
sizes = [200,300,450,350,600,400,300]
alphas = [0.9,0.4,0.67,0.8,0.7,0.9,0.8]
plt.scatter(day,no, color = c,s = sizes,alpha = alphas)
plt.title("Scatter Plot", fontsize = 12, font = "Times New Roman")
plt.xlabel("Day")
plt.ylabel("Number")
plt.show()
```

```python
day = [1,2,3,4,5,6,7]
no = [2,3,1,4,5,3,6]
c = ["yellow","red","blue","green","orange","brown","black"]
plt.scatter(day,no, color = c,s = 500, marker = "*")
plt.scatter(day,no, color = c,s = 300, marker = "P")
plt.scatter(day,no, color = c,s = 200, marker = ">")
plt.scatter(day,no, color = c,s = 200, marker = "<")
plt.title("Scatter Plot", fontsize = 12)
plt.xlabel("Day")
plt.ylabel("Number")
plt.show()
```

```python
day = [1,2,3,4,5,6,7]
no = [2,3,1,4,5,3,6]
c = ["yellow","red","blue","green","orange","brown","black"]
sizes = [200,300,150,350,600,400,300]
plt.scatter(day,no, color = c,s = sizes, edgecolors = "black", linewidth = 3)
plt.title("Scatter Plot", fontsize = 12, font = "Times New Roman")
plt.xlabel("Day")
plt.ylabel("Number")
plt.show()
```

```python
day = [1,2,3,4,5,6,7]
no = [2,3,1,4,5,3,6]
sizes = [200,300,450,350,600,400,300]
color = [10,49,30,29,56,20,30]
plt.scatter(day,no, c = color, s = sizes, cmap = "viridis", alpha = 0.759)
t = plt.colorbar()
t.set_label("Colour Bar", fontsize = 12)

plt.title("Scatter Plot", fontsize = 12)
plt.xlabel("Day")
plt.ylabel("Number")
plt.show()
```

```python
day = [1,2,3,4,5,6,7]
no = [2,3,1,4,5,3,6]
no2 = [3,2,4,5,1,6,2]
sizes = [200,300,450,350,600,400,300]
color = [10,49,30,29,56,20,30]
plt.scatter(day,no, c = color, s = sizes, cmap = "viridis", alpha = 0.759)
plt.scatter(day,no2, c = "r")
t = plt.colorbar()
t.set_label("Colour Bar", fontsize = 12)

plt.title("Scatter Plot", fontsize = 12)
plt.xlabel("Day")
plt.ylabel("Number")
plt.show()
```
