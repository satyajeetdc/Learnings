```python
import matplotlib.pyplot as plt
```

```python
x = [1,2,3,4,5]
y = [3,1,5,2,6]
```

```python
plt.plot(x, y)
plt.title("ABC Pvt. Ltd.", fontsize = 15)
plt.xlabel("Days -------> ")
plt.ylabel("Growth -------> ")
plt.text(1,5.8,"Sample Text", fontsize = 15, style = "italic",bbox = {"facecolor" : "m"})
plt.show()
```
