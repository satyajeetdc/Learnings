```python
x = ["Python","C","C++","Java"]
y = [85,70,60,82]
plt.xlabel("Language", fontsize = 10)
plt.ylabel("Popularity", fontsize = 10)
plt.title("Language Popularity", fontsize = 15)
colours = ["y","b","r","g"]
plt.bar(x,y,width = 0.6, color = colours, align = "center", edgecolor = "black", linewidth = 2, linestyle = ":")
plt.show()
```

```python
from matplotlib import pyplot as plt
import numpy as np

x = ["Python","C","C++","Java"]
y = [85,70,60,82]
z = [70,30,50,60]
p = np.arange(len(x))
width = 0.4
p1 = [(j + width) for j in p]

plt.xlabel("Language", fontsize = 10)
plt.ylabel("Scale", fontsize = 10)
plt.title("Language Popularity", fontsize = 15)

plt.bar(p, y, width, color = "blue", label = "Popularity Index")
plt.bar(p1, z, width, color = "orange", label = "Users in millions")

plt.xticks(p+width/2, x, rotation = 45)
plt.legend()
plt.show()
```

```python
from matplotlib import pyplot as plt
import numpy as np

x = ["Python","C","C++","Java"]
y = [85,70,60,82]
z = [70,30,50,60]
p = np.arange(len(x))
width = 0.4
p1 = [(j + width) for j in p]

plt.xlabel("Scale", fontsize = 10)
plt.ylabel("Language", fontsize = 10)
plt.title("Language Popularity", fontsize = 15)

plt.barh(p, y, width, color = "blue", label = "Popularity")
plt.barh(p1, z, width, color = "orange", label = "Users")

plt.yticks(p+width/2, x, rotation = 0)
plt.legend()
plt.show()
```
