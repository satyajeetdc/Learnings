```python
from matplotlib import pyplot as plt
```

```python
x = [10,20,30,40]
y = ["Cash","CC - Credit Card","DC - Debit Card","UPI"]
```

```python
plt.pie(x, labels = y)
plt.show()
```

```python
ex = [0.15,0.0,0.0,0.0]
plt.pie(x, labels = y, explode = ex)
plt.show()
```

```python
c = ["green","yellow","red","darkblue"]
ex = [0.15,0.0,0.0,0.0]
plt.pie(x, labels = y,colors = c, explode = ex)
plt.show()
```

```python
ex = [0.15,0.0,0.0,0.0]
plt.pie(x, labels = y, explode = ex, autopct = "%0.0f%%")
plt.show()
```

```python
ex = [0.15,0.0,0.0,0.0]
plt.pie(x, labels = y, explode = ex, autopct = "%0.0f%%", shadow = True)
plt.show()
```

```python
ex = [0.15,0.0,0.0,0.0]
plt.pie(x, labels = y, explode = ex, autopct = "%0.0f%%", radius = 0.85, labeldistance = 1.1)
plt.show()
```

```python
ex = [0.15,0.0,0.0,0.0]
plt.pie(x, labels = y, explode = ex, autopct = "%0.1f%%", startangle = 180)
plt.show()
```

```python
ex = [0.15,0.0,0.0,0.0]
plt.pie(x, labels = y, explode = ex, autopct = "%0.0f%%", textprops = {"fontsize" : 12})
plt.show()
```

```python
ex = [0.15,0.0,0.0,0.0]
plt.pie(x, labels = y, explode = ex, autopct = "%0.0f%%", textprops = {"fontsize" : 12}, counterclock = False)
plt.show()
```

```python
ex = [0.15,0.0,0.0,0.0]
plt.pie(x, labels = y, explode = ex, autopct = "%0.0f%%", textprops = {"fontsize" : 12}, 
        shadow = True, wedgeprops = {"linewidth" : 3})
plt.show()
```

```python
ex = [0.15,0.0,0.0,0.0]
plt.pie(x, labels = y, explode = ex, autopct = "%0.0f%%", textprops = {"fontsize" : 12}, 
        shadow = True, wedgeprops = {"linewidth" : 2,"edgecolor" : "m"})
plt.show()
```

```python
ex = [0.15,0.0,0.0,0.0]
plt.pie(x, labels = y, explode = ex, autopct = "%0.0f%%", textprops = {"fontsize" : 12}, 
        wedgeprops = {"linewidth" : 2}, rotatelabels = True)
plt.show()
```

```python
ex = [0.15,0.0,0.0,0.0]
plt.pie(x, labels = y, explode = ex, autopct = "%0.0f%%", textprops = {"fontsize" : 12}, 
        wedgeprops = {"linewidth" : 2}, rotatelabels = False)
plt.title("Payment Types")
plt.show()
```

```python
ex = [0.15,0.0,0.0,0.0]
plt.pie(x, labels = y, explode = ex, autopct = "%0.0f%%", textprops = {"fontsize" : 12}, 
        wedgeprops = {"linewidth" : 2}, rotatelabels = False)
plt.title("Payment Types")
plt.legend(loc = 2)
plt.show()
```

```python
z = [40,30,20,10]
c1 = ["red","blue","green","yellow"]
plt.pie(x, labels = y, colors = c1, radius = 1)
plt.pie(z, colors = "w", radius = 0.6)
plt.title("Payment Types")
plt.show()
```

```python
z = [40,30,20,10]
c1 = ["red","blue","green","yellow"]
plt.pie(x, labels = y, colors = c1, radius = 1)

cr = plt.Circle(xy = (0,0), radius = 0.6, facecolor = "white")
plt.gca().add_artist(cr)

plt.title("Payment Types")
plt.show()
```
