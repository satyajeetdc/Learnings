```python
from matplotlib import pyplot as plt
```

```python
x = [1,2,3,4,6,4,2]
y = [1,2,3,4,6,4,2]
```

```python
plt.subplot(2, 2, 1)
plt.plot(x, y, color = "r")

plt.subplot(2, 2, 2)
plt.pie([1,2], colors = ["green","r"])

plt.subplot(2, 2, 4)
plt.step(x, y)

plt.subplot(2, 2, 3)
plt.bar(x, y)
plt.show()
```
