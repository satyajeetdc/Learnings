```python
from matplotlib import pyplot as plt
```

```python
x = [8,4,2,10,6]
y = [1,10,5,4,2]
```

```python
plt.stem(x,y, linefmt = ":", markerfmt = "r^")
plt.show()
```

```python
plt.stem(x,y, linefmt = ":", markerfmt = "r^", bottom = 5, basefmt = "g", label = "Python")
plt.legend()
plt.show()
```

```python
plt.stem(x,y, linefmt = ":", markerfmt = "r^", bottom = 5, basefmt = "g", label = "Python", use_line_collection = False)
plt.legend()
plt.show()
```

```python
plt.stem(x,y, linefmt = ":", markerfmt = "r^", bottom = 5, basefmt = "g", label = "Python", orientation = "horizontal")
plt.legend()
plt.show()
```
